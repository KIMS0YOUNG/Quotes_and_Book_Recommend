```python
# Colab에서 Selenium 사용하기
!apt-get update
!apt install chromium-chromedriver
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
```

```python
# 스크롤 내리기 함수
def scroll_down(my_url):
  driver.get(my_url)
  
  # 창 가장 크게
  driver.maximize_window() 
  # 처음 페이지 높이 저장
  first_pg_height =  driver.execute_script("return document.body.scrollHeight")

  # 반복
  while True:
    # 맨 밑으로 스크롤
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2) # 페이지 대기

    # 스크롤 다운한 후 페이지 높이
    last_pg_height =  driver.execute_script("return document.body.scrollHeight")

    # 처음 높이랑 마지막 높이랑 같으면 중단
    if first_pg_height == last_pg_height: 
      break 
    first_pg_height=last_pg_height
  print("---------스크롤 끝---------")
```

```python
# 평점별 영화 url 가져오기
title_list = []
for rating in range (8, 11) :
  url = "https://pedia.watcha.com/ko-KR/users/DgwxAeQYNxrMj/contents/movies/ratings/%d"  %rating
  scroll_down(url)
  time.sleep(0.02)
  html = driver.page_source
  soupwatcha = bs4.BeautifulSoup(html, 'html.parser')
  container = (soupwatcha.find("div", {"class":"css-69efdu-Container e1s6qkvu0"}))

# 영화 제목 가져오기  
  for title in range (len(container)):
    title_find = container.find_all("li", {"class":"css-8y23cj"})
    for i in range (len(title_find)):
      title=title_find[i].text
      title_list.append(title)

print(title_list)
```
