```python
from urllib import request
# 스테디셀러 총 1000개. 한 쪽에 20개씩, 총 50쪽
for page_num in range(1, 51):
    # url: excel 파일 저장 링크
    url = f"http://www.yes24.com/24/category/bestsellerExcel?CategoryNumber=001&sumgb=03&PageNumber={page_num}"
    request.urlretrieve(url, "/content/drive/MyDrive/Quotes_and_Recommend/data/book/책목록%d.html" %page_num)
    print("파일 저장 완료")
    book = pd.read_html("/content/drive/MyDrive/Quotes_and_Recommend/data/book/책목록%d.html" %page_num, encoding="cp949")
    print("파일 html로 읽어오기 완료")
    book_df=book[0] # 리스트를 데이터 프레임으로 변환
    book_df.to_csv("/content/drive/MyDrive/Quotes_and_Recommend/data/book/책목록%d.csv" %page_num) # CSV로 저장
    print("끝")
```
